```python
# Multiplication by math module
%%time
torch.matmul(tensor,tensor)
#'@' symbol is also used for matrix muliplication
```

One of the most common errors in deep learning is shaping (3x2) @ (2x3) - works, but (3x2) @ (3x2) wont!


```python
# Shapes for matrix multiplication (not working example) 3x2 @ 3x2
tensor_A = torch.tensor([[1,2],
                        [3,4],
                        [5,6]])
tensor_B = torch.tensor([[6,7],
                        [8,9],
                        [10,11]])
torch.matmul(tensor_A,tensor_B)

```

```python
# Transpose (fixing shape errors)
tensor_A = tensor_A.T
torch.matmul(tensor_A,tensor_B)
```

```python
# Shapes for matrix multiplication (working example) 3x2 @ 2x3


tensor_C = torch.tensor([[1,2],
                        [3,4],
                        [5,6]])
tensor_D = torch.tensor([[6,7,9],
                        [8,9,8]])
# mm is alias for mathmul
torch.mm(tensor_C,tensor_D)
```

#### 2.3.1 Reshaping, stacking, squeezing and unsqueezing tensors

```python
x = torch.arange(1.,10)
x, x.shape

#Add extra dimension (hint: square bracket)
x_reshaped = x.reshape(1,9)
x_reshaped.shape
```

```python
# Change the view - It SHARES the memory with original tensor! (z is different view of x)
z = x.view(1,9)
z,x
```

```python
# Stack tensors on top of each other
x_stacked = torch.stack([x,x,x,x])
x_stacked
```

```python
# Stack tensors on top of each other with dim1  (rearrange it in columns)
x_stacked = torch.stack([x,x,x,x], dim=1)
x_stacked
```

```python
# torch.squeeze -> remove all single dimensions
print(f"Previous tensor {x_reshaped}")
print(f"Previous shape {x_reshaped.shape}")

x_squuezed = x_reshaped.squeeze()

print(f"Squeezed tensor {x_squuezed}")
print(f"Squeezed tensor shape {x_squuezed}")
```

```python
# torch.unsqueeze -> adds a sinmgle dimension at a specific dim
print(f"Previous tensor {x_squuezed}")
print(f"Previous shape {x_squuezed.shape}")

# It can also rearrange the dimmension dim -> dim=0 == row; dim==1 == column
x_unsqueezed = x_squuezed.unsqueeze(dim=0)

print(f"Unsqueezed tensor {x_unsqueezed}")
print(f"Unsqueezed tensor shape {x_unsqueezed.shape}")
```

```python
# torch.permute -> returns a view of the original tensor with its dimensions permuted

x_original = torch.rand(size=(224,224,3)) # [height, width, color_channels(RGB)]

# Permute the original tensor to rearrange the axis order
x_permuted = x_original.permute(2,0,1) # shift axis 0->1, 1->2, 2->0
print(f"Previous shape: {x_original.shape}")
print(f"New shape: {x_permuted.shape}") # color channels [height, width]

```

## 4.Reproducibility (trying to take random out of random)

`start with random number -> tensor operations -> update random numbers to try and make them better representations of the data -> again -> again -> again ->again...`

To reduce the randomness in neural networks and PyTorch come the concept of **random seed**

```python
import torch

# Create 2 random tensors

random_tensor_A = torch.rand(3,3)
random_tensor_B = torch.rand(3,3)

print(f"Random tensor A: \n{random_tensor_A}\n")
print(f"Random tensor B: \n{random_tensor_B}\n")
print(f"Equality: \n{random_tensor_A == random_tensor_B}")

```

```python
# Make some random, but reproducible tenors
RANDOM_SEED = 42

torch.manual_seed(RANDOM_SEED)
random_tensor_C = torch.rand(3,3)

torch.manual_seed(RANDOM_SEED)
random_tensor_D = torch.rand(3,3)

print(f"Random tensor C: \n{random_tensor_C}\n")
print(f"Random tensor D: \n{random_tensor_D}\n")
print(f"Equality: \n{random_tensor_C == random_tensor_D}")

# Torch.manual_seed works only for 1 line of code!!!

```
